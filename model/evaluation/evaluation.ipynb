```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
file_path = 'metrics_nn1.csv'  # Make sure to update the file path if needed
df = pd.read_csv(file_path)

# Plotting
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot loss and validation loss
ax1.plot(df.index, df['loss'], label='Loss', color='tab:blue')
ax1.plot(df.index, df['val_loss'], label='Validation Loss', color='tab:orange')
ax1.set_xlabel('Iterations')
ax1.set_ylabel('Loss')
ax1.legend(loc='upper left')

# Plot mean absolute error and validation mean absolute error
ax2 = ax1.twinx()
ax2.plot(df.index, df['mean_absolute_error'], label='Mean Absolute Error', color='tab:green', linestyle='dashed')
ax2.plot(df.index, df['val_mean_absolute_error'], label='Validation Mean Absolute Error', color='tab:red', linestyle='dashed')
ax2.set_ylabel('Mean Absolute Error')
ax2.legend(loc='upper right')

# Plot learning rate as a bar graph
ax3 = ax1.twinx()
ax3.bar(df.index, df['lr'], label='Learning Rate', color='tab:gray', alpha=0.2, width=0.5)
ax3.set_ylabel('Learning Rate')
ax3.legend(loc='center right')

# Add titles and grid
plt.title('Metrics Over Iterations')
ax1.grid(True)

# Show the plot
plt.show()
```
